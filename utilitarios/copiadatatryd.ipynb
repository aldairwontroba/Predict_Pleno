```python
import os
import shutil
from datetime import datetime
from pathlib import Path

# Caminhos de origem e destino

ano = 2025
base = Path(r"E:\Mercado BMF&BOVESPA\tryd") / str(ano)
# origem = base / "Extraidos INDWIN"   # ajuste aqui se tiver acento: "Extraídos WININD"
origem = base / "Extraidos DOLWDO"

destino = r"C:\Tryd7\workspace\replay\import\WDO&DOL"
# destino = r"C:\Tryd7\workspace\replay\import\WIN&IND"

# Data limite
data_limite = datetime.strptime(f"{ano}-08-01", "%Y-%m-%d")

if not base.exists():
    raise FileNotFoundError(f"A pasta base não existe: {base}")

# Percorre todas as subpastas
for pasta in os.listdir(origem):
    caminho_pasta = os.path.join(origem, pasta)
    
    if os.path.isdir(caminho_pasta):
        try:
            # Tenta converter o nome da pasta em data
            data_pasta = datetime.strptime(pasta, "%Y-%m-%d")
            
            if data_pasta > data_limite:
                print(f"Copiando arquivos da pasta: {pasta}")
                # Copia todos os arquivos dentro da pasta
                for arquivo in os.listdir(caminho_pasta):
                    caminho_arquivo = os.path.join(caminho_pasta, arquivo)
                    if os.path.isfile(caminho_arquivo):
                        shutil.copy2(caminho_arquivo, destino)
        except ValueError:
            # Caso a pasta não tenha nome de data, ignora
            pass

print("✅ Cópia concluída!")
```
